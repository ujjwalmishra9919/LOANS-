```python
for i in cat_col:
    df_cat_clm=pd.get_dummies(df_cat_clm,columns=[i],drop_first=True,dtype=int)
```

# Concat both cat and num cols

```python
df2=pd.concat([df_cat_clm,df_num_clm], axis=1)
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df2.drop("Default", axis=1)
y = df2["Default"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=12)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

arr = ss.fit_transform(X)

df3_scale = pd.DataFrame(arr, columns=X.columns)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(class_weight='balanced')
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
lr.score(X_test,y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score,f1_score, confusion_matrix, classification_report
```

```python
print('P_score', precision_score(y_test, y_pred))
```

```python
print('R_score', recall_score(y_test, y_pred))
```

```python
print('f1_score', f1_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
